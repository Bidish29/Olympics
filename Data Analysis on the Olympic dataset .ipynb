## Plotting the Graph


#### In order to plot the graph let's define a function to plot interactive graphs for the Summer Olympics.

```python
import seaborn as sns

def newplot_summer(Year):
    data = df_new[df_new.Year == Year]
    pp.figure(figsize=(9.5,9.5))
    sns.set_style("whitegrid")
    sns.scatterplot(x='Year',y='Count_Per_Year',hue=df_new['Sport'],data=data,s=150,edgecolor="k")
    
    
    pp.axis(ymin=0,ymax=3000,xmin=1960,xmax=2020)
    pp.xticks(np.arange(1960, 2020, step=4))
    pp.yticks(np.arange(0, 3000, step=100))
    pp.xlabel('Year',fontsize=14, labelpad=10)
    pp.ylabel('Number of Participants',fontsize=14, labelpad=15)
    pp.title('How sports at Summer Olympics have evolved over the past 50 years.',fontsize=16,y=1.02)
    pp.tick_params(labelsize=13, pad=6)
    
```

#### Now, to make an interactive visualization, we will add an interact() fucntion to see how the games have evolved over 50 years.

```python
interact(newplot_summer,Year=widgets.IntSlider(min=1964,max=2016,step=4))
```
